```python
#Libraries
import pandas as pd
import statsmodels.api as sm
import numpy as np
import seaborn as sns
```

```python
#Loading the data
df=pd.read_csv("Clothing.csv")
df.head()
```

#Analyzing Data

```python
#Picking variables
df = df[["tsales", "margin", "nown", "inv1", "inv2", "ssize", "start"]]
df.head(1)
```

```python
#Summary statistics
df.describe()
```

```python
#Plotting IV against DV
sns.pairplot(data=df, y_vars='tsales',x_vars=['margin','ssize'],height=6)
```

```python
#Correlation matrix
sns.heatmap(df.corr(),
            annot=True,
            fmt='.1g',
            cmap='coolwarm',
            linewidth=1,
            linecolor='black'
            )
```

#Training and Test Set

```python

#Isolate x and y
x=df.drop(columns=['tsales'])
y=df.tsales
#Add a constant
x=sm.add_constant(x)
#Training and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,
                                                 test_size=0.2,
                                                 random_state=1502)
df.head(1)
```

#Multilinear Regression

```python
model=sm.OLS(y_train,x_train).fit()
print(model.summary())
```

```python
#predictions
predictions=model.predict(x_test)
predictions[:2]
```

```python
#Accuracy Assessment <If RMSE is higher than MAE, it means it is highly that the dataset has too many outliers>
from sklearn.metrics import mean_absolute_error, mean_absolute_error, mean_squared_error
print("MAE:",round(mean_absolute_error(y_test, predictions),0))
print("RMSE:", round(np.sqrt(mean_squared_error(y_test, predictions)),0))
```
