```python
#Import Libraries
import pandas as pd
import statsmodels.api as sm
import numpy as np
import seaborn as sns
```

```python
#Look at data set
df.head(1)
```

```python
#Transform object in dummies
df=pd.get_dummies(data=df, drop_first=True)
df.head(2)
```

#Multilinear regression preparation

```python
#isolate Xand Y
y=df.salary/1000
X=df.drop(columns=["salary","yrs.service"])
X.head(1)
```

```python
#Add a constant
X=sm.add_constant(X)
X.head()
```

```python
#Training and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X,y,
                                                   test_size=0.2,
                                                   random_state=1502)
X_train.head(2)
```

#Multilinear Regression


```python
#Multilinear regression
model=sm.OLS(y_train, X_train).fit()
print(model.summary())
```

#Accuracy Assessment

```python
#Predict
predictions=model.predict(X_test)
predictions[:4]
```

```python
#Accuracy Assessment
from sklearn.metrics import mean_absolute_error, mean_absolute_error, mean_squared_error
print("MAE:",round(mean_absolute_error(y_test, predictions),1))
print("RMSE:", round(np.sqrt(mean_squared_error(y_test, predictions)),1))
```
